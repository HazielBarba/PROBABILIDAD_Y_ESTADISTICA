```python
asientos = ["asiento1", "asiento2", "asiento3", "asiento4", "asiento5", "asiento6", "asiento7", "asiento8","asiento9","asiento1o" ]
n = len(asientos)
r = 4 # como agrupar

Pr= mt.factorial(n)/mt.factorial(n-r)

print( "Existen ", Pr , " posibles permutaciones sin repetición del total de ",
       n, " asientos"," en grupo de ", r ," en ", r)

print("Aqui las permutaciones")
permutaciones_nombres = list(permutations(asientos, r))

# Imprime las permutaciones
for permutacion in permutaciones_nombres:
    print(permutacion)
```
